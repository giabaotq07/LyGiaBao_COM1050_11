```python
#7
n = int(input())
notprime = bool(list(range(n+3)))
notprime = [False] * (n+3)
notprime[0] = True
notprime[1] = True
for i in range(2,int((n+1)**0.5)+1):
    if notprime[i] == False:
        for j in range(i*i,n+2,i):
            notprime[j] = True
ans = 0 
for i in range(1,int(n ** 0.5)+1):
    if n % i == 0:
        if notprime[i] == False:
            ans = max (ans , i)
        if notprime[n // i] == False:
            ans = max (ans , n // i)
print(ans)
```

```python
#8
def reverse(x):
    Rnum = 0
    while x > 0:
        Rnum = Rnum * 10 + x % 10
        x //= 10
    return Rnum
def check_palindrome(x):
    if x == reverse(x):
        return True
    else:
        return False
n = int(input())
count = 0
while (check_palindrome(n) == False):
    n = n + reverse(n)
    count += 1
print(f"{count} {n}")
```

```python
#9
n = int(input())
def check(x):
    num = [int(0)] * 10
    while (x > 0):
        d = x % 10
        if (num[d] == 1):
            return False
        num[d] = 1
        x //= 10
    return True
i = 1
while (i*i <= n):
    if (check(i*i)):
        print(i*i,end=' ')
    i += 1
```

```python
#30 
def Upper(s : int) -> bool :
    return s>=65 and s<=90
def common(s : int) -> bool :
    return s>=97 and s<=122
def num (s : int) -> bool :
    return s>=48 and s<=57
s = str(input())
U = 0
C = 0
N = 0
for i in s :
    if Upper(ord(i)):
        U += 1
    elif common(ord(i)):
        C += 1
    elif num(ord(i)):
        N += 1
print(U , C , N , end = ' ')
```

```python
#32
s = str(input())
def weak():
    print("Mật khẩu yếu")
if len(s) <= 6:
    weak()
else:
    U = False
    L = False
    D = False
    S = False
    for i in s:
        if i.isupper():
            U = True
        elif i.islower():
            L = True
        elif i.isdigit():
            D = True
        else:
            S = True
    if U and L and D and S:
        print("Mật khẩu mạnh")
    else:
        weak()
```

```python
#34
a = str(input())
b = str(input())
ans = ''
tmp = ''
j = 0
def reset():
    global tmp , j
    tmp = ''
    j = 0
for i in a:
    if i == b[j]:
        tmp += i
        j += 1
        if j == len(b):
            reset()
    elif i == b[0]:
        ans += tmp
        tmp = i
        j = 1
    else:
        ans += tmp + i
        reset()

ans += tmp
print(ans)
```
