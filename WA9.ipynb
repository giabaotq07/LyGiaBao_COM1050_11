```python
#W9A2
def merge(a , b , i ,j , ans):
    if (i == len(a)):
        for k in range(j , len(b)):
            ans.append(b[k])
        return ans
    elif (j == len(b)):
        for k in range(i , len(a)):
            ans.append(a[k])
        return ans
    else:
        if (a[i] < b[j]):
            ans.append(a[i])
            return merge(a , b , i + 1 , j , ans)
        else:
            ans.append(b[j])
            return merge(a , b , i , j + 1 , ans)
k = 0
a = []
while True:
    line = list(map(int , input().split()))
    if not line:
        k = a[len(a) - 1][0]
        a.pop()
        break
    else:
        a.append(line)
ans = []
for row in a:
    ans = merge(ans , row , 0 , 0 ,[])
print(ans[k-1])
```

```python
#W9A3
def change(i ,  j , s):
    if (i == len(s)):
        return s
    minn = int(s[i])
    pos = i
    for j in range(i + 1 , len(s)):
        if (int(s[j]) < minn):
            minn = int(s[j])
            pos = j
    if (minn < int(s[i])):
        s[i] , s[pos] = s[pos] , s[i]
        return s
    else:
        return change(i + 1 , j , s)
s = input()
a =list(s)
a = change(0 , 1 , a)
ans = ''
for i in a:
    if ((not ans)  and (i == '0')):
        continue
    ans += i
print(ans)
```

```python
#W9A7
def canPlaceFlowers(place ,i ,k):
    if (k == 0):
        return True
    if (i == len(place)):
        return False
    if (place[i] == 0):
        checkleft = True
        checkright = True
        if (i != 0) and (place[i - 1] != 0):
            checkleft = False
        if (i != len(place) - 1) and (place[i + 1] != 0):
            checkright = False
        if (checkleft and checkright):
            place[i] = 1
            return canPlaceFlowers(place , i + 1 , k - 1)
    return canPlaceFlowers(place , i + 1 , k)
place = list(map(int , input().split()))
k = int(input())
print(canPlaceFlowers(place , 0 , k))
```

```python
#W9A8
map = {"R" : [1 , 0] , "L" : [-1 , 0] , "U" : [0 , 1] , "D" : [0 , - 1]}
def is_path_crossing(i , moves , x , y , crossed):
    if (crossed[x][y]):
        return True
    if (i == len(moves)):
        return False
    crossed[x][y] = True
    u , v = map[moves[i]]
    return is_path_crossing(i + 1, moves , x + u , y + v , crossed)
moves = input()
crossed = [[False for _ in range(len(moves)*2 + 1)] for _ in range(len(moves)*2 + 1)]
print(is_path_crossing(0 , moves , len(moves) , len(moves) , crossed))
```

```python
#W9A9
def Median(nums):
    a , b , c , d , e  = nums
    def sort(x , y):
        return (x , y) if (x < y) else (y , x)
    a , b = sort(a , b)
    c , d = sort(c , d)
    #2
    if (a > c):
        a , b , c , d = c , d , a , b
    #3
    a = e
    a , b = sort(a , b)
    #4
    if (a > c):
        a , b , c , d = c , d , a , b
    #5
    if (b < c):
        return b
    #6
    return c 
nums = list(map(int,input().split()))
print(Median(nums))
```

```python
#W9A10
def findRadius(i , j ,houses , heaters):
    if (i == len(houses)):
        return 0
    if (houses[i] < heaters[j]):
        d = abs(houses[i] - heaters[j])
        if (j != 0):
            d = min(d , houses[i] - heaters[j - 1])
        return max(d , findRadius(i + 1 , j , houses , heaters))
    if (j == len(heaters) - 1):
        return max(abs(houses[i] - heaters[j]) , findRadius(i + 1 , j , houses , heaters))
    return findRadius(i , j + 1 , houses , heaters)
houses = list(map(int , input().split()))
heaters = list(map(int , input().split()))
houses.sort()
heaters.sort()
print(findRadius(0 , 0 ,houses , heaters))
```
