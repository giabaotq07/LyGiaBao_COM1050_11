```python
#1
def Max_Of_Two(a , b)-> int :
    if a > b :
        return a
    else :
        return b

a , b = map(int , input().split())
print(Max_Of_Two(a, b))
```

```python
#2
def swap(a , b) -> tuple[int , int]:
    return b , a
a , b = map(int , input().split())
c = swap(a , b)
print(c[0] , c[1] , end = ' ')
```

```python
#3
def Isprime(n: int) -> bool:
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True 
n = int(input())
print(Isprime(n))
```

```python
#4
def Isperfect(n: int) -> bool:
    sum = 0
    for i in range(1 , int(n**0.5) + 1):
        if n % i == 0:
            sum += i
            if i != (n // i) :
                sum += n // i
    return sum - n == n
n = int(input())
print(Isperfect(n))
```

```python
#5
def kth_val(a: list[int] , k : int) -> int:
    for i in range(len(a)):
        if (a[i] == k):
            return i+1
    return -1 
a = list(map(int , input().split()))
k = int(input())
print(kth_val(a , k))
```

```python
#6
def fac(n : int) -> int:
    ans = 1
    for i in range(1 , n + 1):
        ans *= i
    return ans
n = int(input())
print(fac(n))
```

```python
#7
def cal(a : float , o : str , b : float) -> float:
    if o == '+' :
        return a + b
    elif o == '-' :
        return a - b
    elif o == '*' :
        return a * b
    else :
        return a / b
a , o , b = map(str , input().split())
print(f"{cal(float(a) , o , float(b)):.2f}")
```

```python
#8
def H_Distance(a: int , b: int) -> int:
    bita = ''
    bitb = ''
    while a > 0 :
        bita = str(a % 2) + bita
        a //= 2
    while b > 0 :
        bitb = str(b % 2) + bitb
        b //= 2
    if len(bita) < len(bitb) :
        bita = '0' * (len(bitb) - len(bita)) + bita
    else :
        bitb = '0' * (len(bita) - len(bitb)) + bitb
    diff = 0
    for i in range(len(bita)) :
        if bita[i] != bitb[i] :
            diff += 1
    return diff

a , b = map(int , input().split())
print(H_Distance(a , b))
```

```python
#10

def build(a : str) -> str:
    vnum = [int(0)] *91
    ans = ''
    for i in range(len(a)):
        if vnum[ord(a[i]) - 32] == 0:
            vnum[ord(a[i]) - 32] = i+1
        ans = ans + str(vnum[ord(a[i]) - 32]) + ' '
    return ans
def check(a: str , b :str) -> bool:
    return build(a) == build(b)
a , b = map(str , input().split())
if check(a , b):
    print("true")
else:
    print("false")
```
