```python
#W7A2
def count_occurences(arr: list , x: int) -> int:
    arr.sort()
    cnt = 0
    tmp = 0
    for i in arr:
        if i == x:
            tmp += 1
        else:
            tmp = 0
        cnt = max(cnt , tmp)
    return cnt 
arr = list(map(int,input().split()))
x = int(input())
print(count_occurences(arr , x))
```

```python
#W7A3 (BUBBLE)
def bubble_sort(arr : list) -> list:
    n = len(arr)
    for k in range(0 , n):
        for i in range(0 , n - 1):
            if (arr[i] > arr[i + 1]):
                arr[i] , arr[i + 1] = arr[i + 1] , arr[i]
    return arr
arr = list(map(int,input().split()))
arr = bubble_sort(arr)
for i in arr:
    print(i,end = ' ')
```

```python
#W7A3 (SELECTION)
def selection_sort(arr : list) -> list:
    n = len(arr)
    for i in range(1,n):
        j = i
        while (j > 0) and (arr[j] < arr[j-1]):
            arr[j] , arr[j-1] = arr[j-1] , arr[j]
            j -= 1
    return arr
arr = list(map(int,input().split()))
arr = selection_sort(arr)
for i in arr:
    print(i,end = ' ')
```

```python
#W7A9
#Prepare
class Fenwicktree:
    def __init__(self , size):
        self.tree = [0] *(size+1)
        self.size = size
    def update(self , i , val):
        i+=1 
        while i <= self.size:
            self.tree[i] += val
            i += i & (-i)
    def query(self , i):
        sum = 0
        i += 1
        while i > 0:
            sum += self.tree[i]
            i -= i &(-i)
        return sum
#init
arr = eval(input())
cnt = Fenwicktree(len(arr))
compress = arr[:] 
ans = []
#compress = list(dict.fromkeys(compress))
#function
def Find_pos(arr , val):
    l = 0
    r = len(arr) - 1
    while (l <= r):
        mid = (l + r)>>1
        if (arr[mid] < val):
            l = mid + 1
        else:
            r = mid - 1
    return l 
#main
compress.sort()
compress = list(dict.fromkeys(compress))
for i in range(0, len(arr)):
    arr[i] = Find_pos(compress , arr[i])
for i in range(len(arr)-1 , -1 , -1):
    cnt.update(arr[i] , 1)
    ans.append(cnt.query(arr[i] - 1))
ans.reverse()
print(ans)

```
