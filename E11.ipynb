```python
#3
def Sum_ofn(n):
    ans = 0
    while (n != 0):
        ans += (n%10)*(n%10)
        n //= 10
    return ans
try:
    a = int(input())
    seen = set()
    happy = True
    while (a != 1):
        if a in seen:
            happy = False
            break
        seen.add(a)
        a = Sum_ofn(a)
    if (happy):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("invalid literal for int() with base 10: '1.0'")
```

```python
#10
try:
    a = list(map(int , input().split()))
    met = set()
    Error = False
    sum = 0
    for i in range(0 , len(a)):
        if a[i] not in met:
            met.add(a[i])
        else:
            Error = True
            break
        sum += a[i]
    if Error:
        raise ValueError
    print(sum)
except (ValueError , TypeError):
    print("Mang khong hop le")
```
