```python
#W8A1
class Cylinder:
    def __init__(self , radius ,height):
        self.radius = radius
        self.height = height
    def surface_area(self):
        return self.radius*2*3.14*self.height + 2*3.14*self.radius**2
    def volume(self):
        return self.radius**2*3.14*self.height
radius , height = map(int , input().split())
c = Cylinder(radius , height)
print(f'{c.surface_area():.2f} {c.volume():.2f}')
```

```python
#W8A3
class point:
    def __init__(self , x , y):
        self.x = x
        self.y = y
    def checkisO(self):
        return self.x == 0 and self.y ==0
    def checkonX(self):
        return self.y == 0
    def checkonY(self):
        return self.x == 0
    def distance(self):
        return ((self.x)**2 + (self.y)**2)**0.5
x,y = map(int , input().split())
d = point(x , y)
print(d.checkisO() , d.checkonX() , d.checkonY())
print(f'{d.distance():.2f}')
```

```python
#W8A5
class shopping_cart:
    def __init__(self):
        self.data = {}
        self.totalmoney = 0
    def add_goods(self , name , price):
        self.data[name] = price
        self.totalmoney += price
    def remove_goods(self , name):
        self.totalmoney -= self.data[name]
        del self.data[name]
    def Empty_now(self):
        return not self.data
    def Totalcost(self):
        return self.totalmoney
    def showall(self):
        for key in self.data:
            print(f'name:{key} , price (dollars) :{self.data[key]}')
    def delall(self):
        self.data.clear()
account = shopping_cart()
stop = 'no'
while (stop != 'yes'):
    print("""
Menu:
1.add a new product 
2.remove a product 
3.your cart is empty now?
4.Total fee you must to pay for all your goods you bought
5.Show all your product
6.Clear your shopping cart
""")
    command = int(input())
    if command == 1:
        name = input("Enter your product's name: ")
        price = int(input("Enter your product's price (dollars): "))
        account.add_goods(name , price)
    elif command == 2:
        name = input("Enter your product's name: ")
        account.remove_goods(name)
    elif command == 3:
        if (account.Empty_now()):
            print("Yes")
        else:
            print("No")
    elif command == 4:
        print(f'Your all cost : {account.Totalcost()}')
    elif command == 5:
        account.showall()
    elif command == 6:
        account.delall()
    print('Do you want to exit now ?')
    stop = input()
```

```python
#W8A6
class rectangle:
    def __init__(self , w , h):
        self.w = w
        self.h = h
    def scale(self , k):
        self.w *= k
        self.h *= k
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return (self.w + self.h) * 2
w , h , k  = map(float , input().split())
figure = rectangle(w , h)
figure.scale(k)
print(f'{figure.area():.2f} {figure.perimeter():.2f}')
```

```python
#W8A7
class fraction:
    def __init__(self , num , den):
        self.num = num 
        self.den = den 
    def divisor_common(self):
        a = self.num
        b = self.den 
        while (b > 0):
            r = a % b
            a = b
            b = r
        return a
    def add(self , num , den):
         self.num = self.num*den + num*self.den
         self.den *= den 
         dc = self.divisor_common()
         self.num = self.num // dc
         self.den = self.den // dc
    def sub(self , num , den):
         self.num = self.num*den - num*self.den
         self.den *= den 
         dc = self.divisor_common()
         self.num = self.num // dc
         self.den = self.den // dc
    def mul(self , num , den):
         self.num *= num
         self.den *= den 
         dc = self.divisor_common()
         self.num = self.num // dc
         self.den = self.den // dc
    def div(self , num , den):
         self.num *= den
         self.den *= num 
         dc = self.divisor_common()
         self.num = self.num // dc
         self.den = self.den // dc
    def __str__(self):
        return str(self.num) + '/' + str(self.den)
a , b , op , c , d = map(str , input().split())
a = int(a)
b = int(b)
c = int(c)
d = int(d)
frac = fraction(a , b)
if op == '+':
    frac.add(c , d)
elif op == '-':
    frac.sub(c , d)
elif op == '*':
    frac.mul(c , d)
elif op == '/':
    frac.div(c , d)
print(frac)
```

```python
#W8A8
class BankAccount:
    def __init__(self , owner , balance):
        self.owner = owner 
        self.balance = balance 
    def withdraw(self , x):
        if x <= self.balance:
            self.balance -= x
    def deposit(self , x):
        self.balance += x 
    def __str__(self):
        return str(f'{self.balance :.2f}')
owner , balance = map(str , input().split())
balance = int(balance)
account = BankAccount(owner , balance)
n = int(input())
for i in range(n):
    command , val = map(str , input().split())
    val = int(val)
    if (command == 'WITHDRAW'):
        account.withdraw(val)
    elif (command == 'DEPOSIT'):
        account.deposit(val)
print(account)
```

```python
#W8A9
class student:
    def __init__(self , name):
        self.name = name
        self.nums = 0
        self.avg = 0
        self.subject = {}
    def add_subject(self , subject , score):
        self.subject[subject] = score
        self.avg = self.avg * self.nums + score
        self.nums += 1
        self.avg /= self.nums
    def avg_grade(self):
        return self.avg
    def rank(self):
        if self.avg >= 8:
            return 'Excellent'
        elif self.avg >= 6.5:
            return 'Good'
        elif self.avg >= 5:
            return 'Average'
        else:
            return 'Poor'
name = input()
achivement = student(name)
n = int(input())
for i in range(n):
    subject , score = map(str , input().split())
    score = float(score)
    achivement.add_subject(subject , score)
print(f'{achivement.name} {achivement.avg_grade():.2f} {achivement.rank()}')    
```

```python
#W8A11
import math 
def len(n):
    return (n[0]**2 + n[1]**2 + n[2]**2)**0.5
def normal(A , B , C):
    AB = []
    for i in range(3):
        AB.append(B[i] - A[i])
    AC = []
    for i in range(3):
        AC.append(C[i] - A[i])  
    return [AB[1]*AC[2] - AB[2]*AC[1] , AB[2]*AC[0] - AB[0]*AC[2] , AB[0]*AC[1] - AB[1]*AC[0]]   
def cosplane(n1 , n2):
    product = 0
    for i in range(3):
        product += n1[i] * n2[i]
    product = abs(product)
    return (product /(len(n1) * len(n2)))
A = list(map(float,input().split()))
B = list(map(float,input().split()))
C = list(map(float,input().split()))
D = list(map(float,input().split()))
n1 = normal(A , B , C)
n2 = normal(B , C , D)
print(f'{math.degrees(math.acos(cosplane(n1 , n2))):.2f}')
```

```python
#W8A12
class Complexnumber:
    def __init__(self , real , img):
        self.real = real
        self.img = img 
    def sum(self , real , img):
        befr = self.real
        befi = self.img
        self.real = befr*real - befi*img
        self.img = befr*img + befi*real
    def __str__(self):
        return str(self.real)+' '+str(self.img)
a , b = map(int , input().split())
c , d = map(int , input().split())
cc = Complexnumber(a , b)
cc.sum(c,d)
print(cc)
```

```python
#W8A13 
class Triangle:
    def __init__(self , a , b , c):
        self.a = a
        self.b = b
        self.c = c
        self.p = (a + b + c)/2
    def getArea(self):
        a = self.a
        b = self.b
        c = self.c
        p = self.p
        if (a + b > c) and (a + c > b) and (b + c > a):
            return (p *(p - a)*(p - b)*(p - c))**0.5
        return -1
a , b ,c = map(int , input().split())
figure = Triangle(a , b , c)
Area = figure.getArea()
if (Area == -1):
    print('invalid')
else:
    print(f'{Area:.2f}')
```
